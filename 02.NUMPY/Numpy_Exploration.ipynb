```python
def initialisation(m,n):
    A = np.random.randn(m,n+1)
    B = np.ones((m,1))
    C = np.concatenate((A,B), axis = 1)
    return C
```

# Travail sur une image

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
plt.imshow(face[int(face.shape[0]/4):-int(face.shape[0]/4), int(face.shape[1]/4):-int(face.shape[1]/4)], cmap=plt.cm.gray)
plt.show()
```
