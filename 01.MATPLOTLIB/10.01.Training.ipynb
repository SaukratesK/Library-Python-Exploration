```python
from collections import Counter

data = pd.read_csv('data.csv')
ids = data['Responder_id']
langage_prog = data['LanguagesWorkedWith']

nouveauCompteur = Counter()

lang = []
popularity = []


for langage in langage_prog:
    nouveauCompteur.update(langage.split(';'))

for i,j in nouveauCompteur.most_common(50):
    lang.append(i)
    popularity.append(j)

print(lang)
print(popularity)

plt.style.use('default')
plt.style.use('dark_background')

fig, axs = plt.subplots(1,2, figsize=(10,12))


lang.reverse()
popularity.reverse()

axs[0].barh(lang, popularity)

lang.reverse()
popularity.reverse()

axs[1].plot(popularity, lang)


```
